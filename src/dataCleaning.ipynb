```python
import json
import pandas as pd
```

```python
with open('json files/match_results.json') as f:
    data = json.load(f)
df_matches = pd.DataFrame(data[0]['matchSummary'])
df_matches.head()    
```

```python
df_matches.shape
```

```python
df_matches.rename({'scorecard' : 'Match ID'}, axis = 1, inplace = 1)
df_matches.head()
```

```python
df_matches.to_csv( 'csv files/match_results.csv', index = False)
```

```python
match_ids_dict = {}

for index, row in df_matches.iterrows():
    key1 = row['team1'] + ' Vs ' + row['team2']
    key2 = row['team2'] + ' Vs ' + row['team1']
    
    match_ids_dict[key1] = row['Match ID']
    match_ids_dict[key2] = row['Match ID']

match_ids_dict    
```

```python
with open('json files/bowling_summary.json') as bowl:
    data1 = json.load(bowl)
    bowl_records=[]
    for i in data1:
        bowl_records.extend(i['bowlingSummary'])

df_bowl = pd.DataFrame(bowl_records)
df_bowl
```

```python
df_bowl["Match ID"] = df_bowl["match"].map(match_ids_dict)
df_bowl.head(10)
```

```python
df_bowl.to_csv("csv files/bowling_summary.csv", index = False)
```

```python
with open('json files/batting_summary.json') as bat:
    data2=json.load(bat)
    all_records = []
    for i in data2:
        all_records.extend(i['battingSummary'])

df_bat = pd.DataFrame(all_records) 
df_bat.head(11)
```

```python
df_bat['out/not_out'] = df_bat.dismissal.apply(lambda x: "out" if len(x)>0 else "not_out")
df_bat.head(11)
```

```python
df_bat['Match ID'] = df_bat['match'].map(match_ids_dict)
df_bat.head()
```

```python
df_bat.drop(columns=["dismissal"], inplace=True)
df_bat.head(10)
```

```python
df_bat['batsmanName'] = df_bat['batsmanName'].apply(lambda x: x.replace('â€', ''))
df_bat['batsmanName'] = df_bat['batsmanName'].apply(lambda x: x.replace('\xa0', ''))
df_bat['batsmanName'] = df_bat['batsmanName'].apply(lambda x: x.replace('(c)', ''))
df_bat.head()
```

```python
df_bat.to_csv("csv files/batting_summary.csv", index = False)
```

```python
with open('json files/player_details.json') as pl:
    data3 = json.load(pl)
df_pl = pd.DataFrame(data3)
df_pl['name'] = df_pl['name'].apply(lambda x : x.replace('â€', ''))
df_pl['name'] = df_pl['name'].apply(lambda x : x.replace('†', ''))
df_pl['name'] = df_pl['name'].apply(lambda x : x.replace('\xa0', ''))
df_pl.head(10)
```

```python
df_pl.to_csv("csv files/players_details.csv", index = False)
```
